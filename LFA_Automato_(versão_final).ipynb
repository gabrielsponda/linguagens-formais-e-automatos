# Desenvolvimento de um Autômato Finito para Validação de Senhas

Para desenvolver um validador de senha eficaz que confirme se uma senha possui ao menos oito caracteres, incluindo uma letra maiúscula, uma letra minúscula, um número e um caractere especial, seguimos uma metodologia estratégica e modular. Esta abordagem foi necessária dada a complexidade de criar um único autômato finito que atendesse a todos esses critérios simultaneamente. Abaixo, descrevemos detalhadamente o procedimento adotado, passo a passo, para a construção e a verificação do autômato, fornecendo insights importantes para a elaboração do relatório.

## Passo 1: Autômatos Separados para Comprimento e Diversidade de Caracteres

- **Comprimento da Senha**: Construímos um autômato inicial para verificar o critério de comprimento mínimo das senhas. Esse autômato percorre uma sequência de estados que representam cada caractere lido até satisfazer o requisito mínimo de oito caracteres.

- **Diversidade de Caracteres**: Desenvolvemos um segundo autômato destinado a assegurar a presença de ao menos um caractere de cada tipo necessário: uma letra maiúscula, uma letra minúscula, um número e um caractere especial. Esse autômato avança através de estados que indicam a detecção de cada tipo de caractere, culminando em um estado que confirma a presença de todos os tipos exigidos.

## Passo 2: Simplificação da Entrada através de Conversão de Caracteres

Para simplificar o processo, implementamos uma função de conversão que transforma a entrada de senhas, categorizando cada caractere em um representante único (`a` para letras minúsculas, `A` para maiúsculas, `1` para números e `#` para caracteres especiais). Essa estratégia minimiza a complexidade do autômato, preservando a validade da verificação ao reduzir a necessidade de considerar o alfabeto completo.

Integrar os dois autômatos (comprimento e diversidade de caracteres) foi o próximo passo, utilizando uma função em Python que aplica o produto entre eles. Essa integração cria um autômato combinado capaz de monitorar ambos os critérios de validação simultaneamente.

## Simplificação da Entrada através de Conversão de Caracteres

```python
def categorizar_entrada(entrada):
    entrada_categorizada = ""
    for char in entrada:
        if char.islower():  # Checa se é minúscula
            entrada_categorizada += 'a'
        elif char.isupper():  # Checa se é maiúscula
            entrada_categorizada += 'A'
        elif char.isdigit():  # Checa se é dígito
            entrada_categorizada += '1'
        else:  # Assume que o restante são caracteres especiais
            entrada_categorizada += '#'
    return entrada_categorizada
```

```python
def afd(M, w):
  # 'extrai' os elementos da tupla
  δ, q, F = M

  for s in w:
    q = δ[(q, s)]

  return q in F
```

```python
def produto_automatos(M, N):
    δ_M, q0_M, F_M = M
    δ_N, q0_N, F_N = N

    # Estado inicial do autômato produto
    q0_produto = (q0_M, q0_N)

    # Conjunto de estados de aceitação do autômato produto
    F_produto = [(q_M, q_N) for q_M in F_M for q_N in F_N]

    # Função de transição do autômato produto
    δ_produto = {}

    # Para cada combinação de estados em M e N
    for ((estado_M, simbolo), prox_estado_M) in δ_M.items():
        for ((estado_N, _), prox_estado_N) in δ_N.items():
            if simbolo == _:  # Garante que o símbolo seja o mesmo para ambas as transições
                estado_atual_produto = (estado_M, estado_N)
                prox_estado_produto = (prox_estado_M, prox_estado_N)
                δ_produto[(estado_atual_produto, simbolo)] = prox_estado_produto

    return δ_produto, q0_produto, F_produto

# Criação do autômato produto
δ_produto, q0_produto, F_produto = produto_automatos(M, N)

# AFD produto representado por uma 3-upla
P = (δ_produto, q0_produto, F_produto)
```

```python
def automata_product_dfa(M, N):
    # Novo conjunto de estados: produto dos conjuntos de estados,
    # convertendo cada par de estados em uma string formatada
    Q_new = {f"{q1},{q2}" for q1 in M.Q for q2 in N.Q}

    # Novo alfabeto: união dos alfabetos (que, neste caso, são iguais)
    sigma_new = M.sigma.union(N.sigma)

    # Nova função de transição
    delta_new = {}
    for q1 in M.Q:
        for q2 in N.Q:
            q_new = f"{q1},{q2}"
            delta_new[q_new] = {}
            for a in sigma_new:
                delta_new[q_new][a] = f"{M.delta[q1][a]},{N.delta[q2][a]}"

    # Novo estado inicial
    initialState_new = f"{M.initialState},{N.initialState}"

    # Novo conjunto de estados finais
    F_new = {f"{q1},{q2}" for q1 in M.F for q2 in N.F}

    # Cria o novo DFA
    new_dfa = DFA(Q_new, sigma_new, delta_new, initialState_new, F_new)
    return new_dfa
```
